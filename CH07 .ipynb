```python
import pandas as pd
url="https://raw.githubusercontent.com/wesm/pydata-book/master/ch02/movielens/movies.dat"
movies=pd.read_table(url,sep="::",encoding = "ISO-8859-1",engine='python',header=None,names=mnames)
```

```python
#First, we extract the list of unique genres in the dataset (using a nice set.union trick):
genre_iter = (set(x.split('|')) for x in movies.genres)
# break a large string down into smaller chunks, or strings. 
genre_iter
```

```python
genres = sorted(set.union(*genre_iter))
```

```python
#one way to construct the indicator DataFrame is to start with a DataFrame of all zeros:
dummies = DataFrame(np.zeros((len(movies), len(genres))), columns=genres)
dummies[:5]
```

```python
# iterate through each movie and set entries in each row of dummies to 1:
# enumerate() returns a tuple containing a count (from start which defaults to 0), and
# the values obtained from iterating over iterable
for i, gen in enumerate(movies.genres):
    dummies.loc[i, gen.split('|')] = 1
# loc[row,column] to select in DataFrame
```

```python
# combine this with movies:
dummies=dummies.add_prefix('Genre_')[0:5]
```

```python
movie_dic=movies.join(dummies)
movie_dic
```

```python
url='https://raw.githubusercontent.com/wesm/pydata-book/master/ch07/foods-2011-10-03.json'
import json
import pandas as pd
from pandas import Series, DataFrame
```

```python
db = json.load(open('foods-2011-10-03.json'))
```

```python
len(db)
```

```python
db[6000]
```

```python
nutrients = DataFrame(db[0]['nutrients'])
nutrients[:7]
```

```python
info_keys = ['description', 'group', 'id', 'manufacturer']
```

```python
info_keys = ['description', 'group', 'id', 'manufacturer']
info = DataFrame(db, columns=info_keys)
info[:3]
```

```python
pd.value_counts(info.group)[:10]
```

```python
nutrients=[]
for rec in db:
    fnuts=DataFrame(rec['nutrients'])
    fnuts['id']=rec['id']
    nutrients.append(fnuts)
nutrients=pd.concat(nutrients,ignore_index=True)
```

```python
nutrients[:2]
```

```python
nutrients.duplicated().sum()
```

```python
nutrients = nutrients.drop_duplicates()
```

```python
col_mapping = {'description' : 'food',  'group' : 'fgroup'}
```

```python
info = info.rename(columns=col_mapping, copy=False)
```

```python
info[:2]
```

```python
col_mapping = {'description' : 'nutrient', 'group' : 'nutgroup'}
nutrients = nutrients.rename(columns=col_mapping, copy=False)
nutrients[:2]
```

```python
 ndata = pd.merge(nutrients, info, on='id', how='outer')
```

```python
ndata[:10]
ndata.dropna()
```

```python
## really save my data!!! pd.pivot_table!!!
pivoted =pd.pivot_table(ndata,index='food', columns='nutrient', values='value')
pivoted[:5]
```

```python
by_nutrient = ndata.groupby(['nutgroup', 'nutrient'])
get_maximum = lambda x: x.xs(x.value.idxmax()) 
get_minimum = lambda x: x.xs(x.value.idxmin())
max_foods = by_nutrient.apply(get_maximum)[['value', 'food']]
max_foods.food = max_foods.food.str[:50]
```
